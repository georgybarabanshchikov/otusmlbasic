```python
from random import randint

import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt

#Сгенерируйте случайную квадратную матрицу A ∈Rn×n, n ≥100
n = 100
minvalue = 0
maxvalue = 65535
mat = np.random.randint(maxvalue, size=(n, n))
# С помощью функции numpy.linalg.svd выполните сингулярное разложение этой матрицы, и получите три матрицы: U , S, V ⊤.
U, s, VT = la.svd(mat)
Sigma = np.diag(s)
# Выполняйте аппроксимацию матрицы A с рангом r, меняя его значение, например, от 2 до n
graph = []
for r in range(1,n):
  graph.append(np.sum((mat - (U[:, :r] @ Sigma[:r, :r] @ VT[:r, :]))**2))

plt.plot(graph)
```
